성별, 연령대 분류로 요일 시간대를 통합

```python
#남성
for i in range(1,7):
    liv_popul['남성 연령대 ' + str(i*10)] = liv_popul.iloc[:,42*i-14:42*i+28].sum(axis=1)
liv_popul
```

```python
#여성
for i in range(1,7):
    liv_popul['여성 연령대 ' + str(i*10)] = liv_popul.iloc[:,42*i+238:42*i+280].sum(axis=1)
    print(i)
liv_popul
```
